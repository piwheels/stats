```python
import matplotlib.pyplot as plt
```

```python
archs = {
    'armv7l': 1480452,
    'armv6l':  142991,
    'x86_64':   51643,
}

arch_pie = plt.pie(archs.values(), labels=archs.keys())
print(archs['armv7l'] / sum(archs.values()))
print(archs['armv6l'] / sum(archs.values()))
print(archs['armv7l'] / sum([archs['armv7l'], archs['armv6l']]))
```

```python
oss = {
    'raspbian': 1588979,
    'debian': 54593,
    'ubuntu': 27220,
    'other': 9374,
}

os_pie = plt.pie(oss.values(), labels=oss.keys())
oss['raspbian'] / sum(oss.values())
```

```python
py_vers = {
    '3.5': 1231797, 
    '2.7': 304622,
    '3.7': 101492,
    '3.6': 38573,
    '3.4': 4740,
    '3.8': 161,
}

py_ver_pie = plt.pie(py_vers.values(), labels=py_vers.keys())
py_vers['3.5'] / sum(py_vers.values())
```

```python
downloads = {
    'April': 626382,
    'May': 535059,
    'June': 520109,
}

downloads_bar = plt.bar(range(len(downloads)), downloads.values(), tick_label=list(downloads.keys()), )
plt.title('Downloads by month')
```

```python
time_saved = {
    'April': 8 + (10.5/12),
    'May': 7 + (11.5/12),
    'June': 7 + (9.5/12),
}

time_saved_bar = plt.bar(range(len(time_saved)), time_saved.values(), tick_label=list(time_saved.keys()))
plt.title('Time saved by month')
plt.ylabel('Years')
```
